```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
post_data = pd.read_csv('posts.csv')
seller_data = pd.read_csv('sellers.csv')
seller_data
```

```python
post_data['Class'].value_counts().plot(kind='bar', color=['red', 'green', 'blue', 'black', 'cyan', 'orange'])
plt.xlabel('Class', labelpad=14)
plt.ylabel('Count', labelpad=14)
plt.title('Post classes', y=1.02)
```

```python
post_data['Category'].value_counts().plot(kind='bar', color=['red', 'green', 'blue', 'black', 'yellow', 'cyan', 'purple', 'orange', 'pink'])
plt.xlabel('Category', labelpad=14)
plt.ylabel('Count', labelpad=14)
plt.title('Post categories', y=1.02)
```
