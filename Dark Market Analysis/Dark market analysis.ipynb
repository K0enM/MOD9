http://darkodemard3wjoe63ld6zebog73ncy77zb2iwjtdjam4xwvpjmjitid.onion. 
(We accessed this market via the Tor Browser that we downloaded). 

Dark0de is an online black market in which all sorts of products are offered by sellers, differing in their legality. As an example, one is able to buy (illegal) drugs or (illegal) software for cheap prices, which is something that is not possible in normal markets. In Dark0de, similar to normal online markets, buyers are able to leave reviews to sellers and rate their interactions with them on a 5-star scale. Based on these reviews and ratings from the public, Dark0de also assigns a symbol to their sellers, ranging from Basic - Gold - Diamond, where diamond means a recognised and high quality seller and Basic means a relatively unrecognised seller by the platform.
The reason that sellers and buyers on black markets, as opposed to normal markets, are able to make these previously mentioned potential illegal transactions is because black markets make use of ‘.onion’ links, which are top level Internet domains used by anonymous websites on the Dark Web. These ‘.onion’ links are only accessible via special browsers that support these links, namely the Tor Browser, which user the Tor protocol under the hood. The ‘.onion’ links in combination with the Tor browser that supports them allow the buyer and sellers to remain anonymous, in turn providing them the possibility to engage in illegal trade and exchange

# Motivation & Goals

In black markets, as mentioned before, a wide variety of products are offered by sellers, of which some are illegal by definition, some are legal and from some products their legality depends. 
In this report, the goals were to investigate how exactly a black market (Dark0de) works in terms of its practice (also see introduction), the range of items that are offered by sellers and to investigate the (il)legality of these offered products.


This investigation into the extent of the (il)legality of the offered products by sellers is important, as this aids law enforcement bodies to subsequently and adequately take the appropriate action against it. Sellers and buyers on black markets are offered the possibility, due to the internet domain on the dark web and the browser that supports it, to remain anonymous while engaging in trade. However, there are certain ways to still trace them and thus to circumvent this anonymity, which is something that is done by law enforcement bodies. However, it should be noted that appropriate action from these bodies requires a thorough investigation into the illegality on behalf of both the seller and the buyer, as this is often a grey area, mostly dependent on the intentions of the parties. 

As an example, a seller might sell a certain software, which is not illegal per se, that is used for protecting the privacy of one’s email (ProtonMail), which in turn is not illegal per se. What if, however, the software is used for sending out phishing emails to innocent people that captures their personal bank data? Then the buyer/user of this software is guilty of illegality in their actions and the law enforcement bodies are then allowed and should take actions against it. This example illustrates the grey area mentioned before and the importance to thoroughly investigate the interaction (specifically the products and the parties’ intentions) between seller and buyer on the black markets.

To do this, the goals combine together to fulfill this goal. First we need to know how the black market (Dark0de) works (goal 1) in order to familiarise ourselves with the range of products offered by sellers (goal 2) and lastly, when this range is clear, superficially investigate the (il)legality of the offered products (often dependent on intentions of the seller and buyer) (goal 3).

## Data analysis

```python
fig, axs = plt.subplots(1, 2, figsize=(12,12))

axs[0] = post_data['Class'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[0])
axs[0].set_title('Post classes')
axs[1] = post_data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[1])
axs[1].set_title('Post Categories')

plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,12))
axs[0] = seller_data['Kind of vendor'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[0])
axs[0].set_title('Vendor kinds')
axs[1] = seller_data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[1])
axs[1].set_title('Seller categories')

plt.show()
```

```python
fig, ax = plt.subplots()
categories = post_data.groupby('Category')

for name, group in categories:
    ax.scatter(group['Times sold'], group['Price (in dollars)'], label=name)
    ax.set_xlabel('Times sold')
    ax.set_ylabel('Price (in dollars)')
    ax.set_title('Price vs Times sold')
    ax.grid(True)
    
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, figsize=(10,10))

categories = seller_data.groupby('Category')

for name, group in categories:
    ax[0].scatter(group['Number of reviews'], group['grade (out of 5)'], alpha=0.6, label=name)
    ax[0].set_xlabel('Number of reviews')
    ax[0].set_ylabel('Rating')
    ax[0].set_title('Number of reviews & ratings')
    ax[0].grid(True)
    
    ax[1].scatter(group['Number of reviews'], group['Active listings'], alpha=0.6, label=name)
    ax[1].set_xlabel('Number of reviews')
    ax[1].set_ylabel('Active listings')
    ax[1].set_title('Number of reviews & active listings')
    ax[1].grid(True)

plt.legend() 
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(13,10))
axs[0,0] = post_data['Times sold'].plot.line(ax=axs[0,0])
post_data['Times sold'].plot.box(ax=axs[0,1],vert=False)
axs[0,0].set_title('Times sold frequency')
axs[0,0].set_xlabel('Times sold')
axs[0,0].set_ylabel('Frequency')
axs[0,1].set_title('Times sold boxplot')
axs[0,1].set_xlabel('Times sold')
post_data['Price (in dollars)'].plot.line(ax=axs[1,0])
post_data['Price (in dollars)'].plot.box(ax=axs[1,1], vert=False)
axs[1,0].set_title('Price (in dollars) frequency')
axs[1,1].set_title('Price (in dollars) boxplot')
axs[1,0].set_xlabel('Price (in dollars)')
axs[1,0].set_ylabel('Frequency')
axs[1,1].set_xlabel('Price (in dollars)')
plt.show()
```
