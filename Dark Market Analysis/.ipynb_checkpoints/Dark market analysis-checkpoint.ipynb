# Introduction

For the course Introduction to Digital forensics we are going to investigate a black market. The black market we are going to investigate in this report is the dark0de market: 

http://darkodemard3wjoe63ld6zebog73ncy77zb2iwjtdjam4xwvpjmjitid.onion. 
(We accessed this market via the Tor Browser that we downloaded). 

Dark0de is an online black market in which all sorts of products are offered by sellers, differing in their legality. As an example, one is able to buy (illegal) drugs or (illegal) software for cheap prices, which is something that is not possible in normal markets. In Dark0de, similar to normal online markets, buyers are able to leave reviews to sellers and rate their interactions with them on a 5-star scale. Based on these reviews and ratings from the public, Dark0de also assigns a symbol to their sellers, ranging from Basic - Gold - Diamond, where diamond means a recognised and high quality seller and Basic means a relatively unrecognised seller by the platform.
The reason that sellers and buyers on black markets, as opposed to normal markets, are able to make these previously mentioned potential illegal transactions is because black markets make use of ‘.onion’ links, which are top level Internet domains used by anonymous websites on the Dark Web. These ‘.onion’ links are only accessible via special browsers that support these links, namely the Tor Browser, which uses the Tor protocol under the hood. The ‘.onion’ links in combination with the Tor browser that supports them allow the buyer and sellers to remain anonymous, in turn providing them the possibility to engage in illegal trade and exchange

# Motivation & Goals

In black markets, as mentioned before, a wide variety of products are offered by sellers, of which some are illegal by definition, some are legal and from some products their legality depends. 
In this report, the goals were to investigate how exactly a black market (Dark0de) works in terms of its practice (also see introduction), the range of items that are offered by sellers and to investigate the (il)legality of these offered products.


This investigation into the extent of the (il)legality of the offered products by sellers is important, as this aids law enforcement bodies to subsequently and adequately take the appropriate action against it. Sellers and buyers on black markets are offered the possibility, due to the internet domain on the dark web and the browser that supports it, to remain anonymous while engaging in trade. However, there are certain ways to still trace them and thus to circumvent this anonymity, which is something that is done by law enforcement bodies. However, it should be noted that appropriate action from these bodies requires a thorough investigation into the illegality on behalf of both the seller and the buyer, as this is often a grey area, mostly dependent on the intentions of the parties. 

As an example, a seller might sell a certain software, which is not illegal per se, that is used for protecting the privacy of one’s email (ProtonMail), which in turn is not illegal per se. What if, however, the software is used for sending out phishing emails to innocent people that captures their personal bank data? Then the buyer/user of this software is guilty of illegality in their actions and the law enforcement bodies are then allowed and should take actions against it. This example illustrates the grey area mentioned before and the importance to thoroughly investigate the interaction (specifically the products and the parties’ intentions) between seller and buyer on the black markets.

To do this, the goals combine together to fulfill this goal. First we need to know how the black market (Dark0de) works (goal 1) in order to familiarise ourselves with the range of products offered by sellers (goal 2) and lastly, when this range is clear, superficially investigate the (il)legality of the offered products (often dependent on intentions of the seller and buyer) (goal 3).

## Data Collection

To get a better picture of this kind of trading we entered the black market Dark0de via the Tor Browser. Once we got access to this market, we collected information from the best sold items of each category, using the filter option and filtering the products based on ‘highest rated’. This information consists of what product was sold, the link/url, times the product was sold, the price of the product and the seller of the product.

After we collected the data of 100 products, we categorized them into whether these sales are ‘legal’, ‘illegal’ or ‘it depends on the situation and intentions’ (also see previous section on motivation and goals).


After we categorized all the sales, we started collecting all available data from the sellers of these 100 products. The collected data of sellers includes: number of reviews, the average rating of the reviews, the number of active listings, the kind of vendor (Basic, Gold or Diamond) and the category in which they had the most products. All data has been put in Google Spreadsheets manually.

## Data analysis

```python
fig, axs = plt.subplots(1, 2, figsize=(12,12))

axs[0] = post_data['Class'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[0])
axs[0].set_title('Post classes')
axs[1] = post_data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[1])
axs[1].set_title('Post Categories')

plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,12))
axs[0] = seller_data['Kind of vendor'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[0])
axs[0].set_title('Vendor kinds')
axs[1] = seller_data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[1])
axs[1].set_title('Seller categories')

plt.show()
```

```python
fig, ax = plt.subplots()
categories = post_data.groupby('Category')

for name, group in categories:
    ax.scatter(group['Times sold'], group['Price (in dollars)'], label=name)
    ax.set_xlabel('Times sold')
    ax.set_ylabel('Price (in dollars)')
    ax.set_title('Price vs Times sold')
    ax.grid(True)
    
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, figsize=(10,10))

categories = seller_data.groupby('Category')

for name, group in categories:
    ax[0].scatter(group['Number of reviews'], group['grade (out of 5)'], alpha=0.6, label=name)
    ax[0].set_xlabel('Number of reviews')
    ax[0].set_ylabel('Rating')
    ax[0].set_title('Number of reviews & ratings')
    ax[0].grid(True)
    
    ax[1].scatter(group['Number of reviews'], group['Active listings'], alpha=0.6, label=name)
    ax[1].set_xlabel('Number of reviews')
    ax[1].set_ylabel('Active listings')
    ax[1].set_title('Number of reviews & active listings')
    ax[1].grid(True)

plt.legend() 
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(13,10))
axs[0,0] = post_data['Times sold'].plot.line(ax=axs[0,0])
post_data['Times sold'].plot.box(ax=axs[0,1],vert=False)
axs[0,0].set_title('Times sold frequency')
axs[0,0].set_xlabel('Times sold')
axs[0,0].set_ylabel('Frequency')
axs[0,1].set_title('Times sold boxplot')
axs[0,1].set_xlabel('Times sold')
post_data['Price (in dollars)'].plot.line(ax=axs[1,0])
post_data['Price (in dollars)'].plot.box(ax=axs[1,1], vert=False)
axs[1,0].set_title('Price (in dollars) frequency')
axs[1,1].set_title('Price (in dollars) boxplot')
axs[1,0].set_xlabel('Price (in dollars)')
axs[1,0].set_ylabel('Frequency')
axs[1,1].set_xlabel('Price (in dollars)')
plt.show()
```

# Conclusions & reflections

As can be seen in several graphs retrieved from the research results, drugs are sold the most on this market. In the second pie chart it can be seen that 19% of the products we have investigated are drugs, which is also the biggest slice of the pie. More specifically scrutinized, of all the dark market sellers identified in this investigation, 23.2% were labelled diamond sellers by Dark0de, meaning they are recognised as being of high-quality and reliable. Next to that, of all the sellers, 32.1% were mainly selling drugs. Lastly, of all the products that were scrutinized in this report, the legality of 40% of these products lied in the grey area, depending on the intentions/context of the exchange of these products.
Another thing that stood out was the fact that there are a lot of data breaches listed. When analyzing this content not a lot of reviews about these products were found. There were significantly few reviews per post compared to, for example, posts for drugs or other services. This could be an indication of disinterest by the customers. There are a lot of offers but very little interest. A second approach is that it could be that the kind of people that buy data breaches are not likely to leave a review. Then the whole idea of high offer and low interest does not hold anymore. 

What is the most important takeaway message/conclusion from the results obtained by this investigation is that the intentions of both the buyer and seller and more generally the ‘context’ is very important for understanding the legality of the interaction between a seller and a buyer on the dark market. In taking account of this ‘context’, perceptions and interpretations are crucial. Where one does not see any illegality in a certain product being bought and exchanged via a black market, another person sees a possibility to misuse the product for illegal activity. This again emphasizes the importance of thorough investigation into the trade and exchange of products on black markets between sellers and buyers on behalf of law enforcement bodies, as they do not want to harass innocent people nor do they want to let people guilty of illegal activity get away.

When reflecting back on the process of our investigation, it also became clear to us that taking into account all these potential interpretations and perceptions towards some product exchange is a tough task. One is ought to be very creative and try to imagine all certain types of possibilities of how a product can be used in order to discover the illegal possibilities as well as the legal possibilities of such product usage. Even if one is almost certainly guilty of illegal action (e.g. DDOS), he/she might even do this while having obtained the consent of the ones he/she is attacking. This again emphasises the importance of the ‘context’, consisting of the intentions/perceptions and possible consent, on determining the legality of exchanges on these black markets, which is often very dependent on the aforementioned factors.
